```python
import yfinance as yf
import pandas as pd
import openpyxl

# Load the Excel file with symbols
file_path = 'symbolsup.xlsx'  # Update with the correct file path
symbols_df = pd.read_excel(file_path)

# Initialize an empty DataFrame to store the data for all stocks
all_data = pd.DataFrame()

# Counter to limit the loop during testing (remove or adjust as needed)
counter = 0

# Iterate over each symbol in the Excel file
for index, row in symbols_df.iterrows():
  try:
    if counter == 10000:  # Adjust the limit as needed
        break
    counter += 1
    symbol = row[1]  # Assuming the symbol is in the second column

    # Fetch the financial data from yfinance
    stock = yf.Ticker(symbol)

    balance_sheet = stock.balance_sheet.T  # Transpose to have dates as rows
    income_statement = stock.financials.T  # Transpose to have dates as rows
    cash_flow = stock.cashflow.T  # Transpose to have dates as rows

    # Combine the three dataframes
    combined_df = pd.concat([balance_sheet, income_statement, cash_flow], axis=1)

    # Add the symbol as the first column
    combined_df.insert(0, 'Symbol', symbol)

    # Add the year as the second column
    combined_df.insert(1, 'Year', combined_df.index.year)

    # Append the combined data to the main DataFrame
    all_data = pd.concat([all_data, combined_df], ignore_index=True)
  except Exception as e:
    print(e)

# Save the combined data for all stocks to an Excel file
all_data.to_excel('combined_financial_data_all_stocks.xlsx', index=False)

print("Data for all stocks saved successfully.")
```
